```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
df = pd.DataFrame(dict_faturamento)
df['data_ref'] = pd.to_datetime(df['data_ref'])


media_vendas = df['valor'].mean()
print(f'Média das vendas: {media_vendas}')
sns.set(style="whitegrid")
```

```python
# Gráfico de barras vertical
plt.figure(figsize=(10, 6))
sns.barplot(x=df['data_ref'].dt.strftime('%Y-%m'), y=df['valor'], palette='viridis')
plt.xlabel('Data de Referência')
plt.ylabel('Valor')
plt.title('Faturamento Mensal')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Gráfico de linhas
plt.figure(figsize=(10, 6))
sns.lineplot(x=df['data_ref'].dt.strftime('%Y-%m'), y=df['valor'], marker='o', color='b')
plt.xlabel('Data de Referência')
plt.ylabel('Valor')
plt.title('Faturamento Mensal')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
